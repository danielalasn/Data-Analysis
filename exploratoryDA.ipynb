```python
import pandas as pd
import numpy as np
import ssl
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline
```

```python
ssl._create_default_https_context = ssl._create_unverified_context
path='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv'
df = pd.read_csv(path)
df
```

```python
print(df.dtypes)
```

```python
df.corr()
```

```python
# Relationship between Engine size and price
print(df[["engine-size", "price"]].corr())
sns.regplot(x='engine-size', y='price', data=df)
plt.ylim(0,)
plt.show()
```

```python
# Relationship between highway mpg and price
print(df[["highway-mpg", "price"]].corr())
sns.regplot(x='highway-mpg', y='price', data=df)
plt.ylim(0,)
plt.show()
```

```python
sns.boxplot(x='body-style', y='price', data=df)
```

```python
sns.boxplot(x="drive-wheels", y="price", data=df)
```

```python
# Descriptive Statistical Analysis
df.describe()
```

```python
df.describe(include=['object'])
```

```python
drive_wheels_counts = df['drive-wheels'].value_counts().to_frame()
drive_wheels_counts.rename(columns={'drive-wheels': 'value_counts'}, inplace=True)
drive_wheels_counts.index.name = 'drive-wheels'
drive_wheels_counts
```

```python
# Grouping
df_group = df[['drive-wheels','body-style','price']]
```

```python
df_group_wheels = df_group.groupby(['drive-wheels'],as_index=False).mean()
df_group_wheels
```

```python
df_group_body = df_group.groupby(['body-style'],as_index=False).mean()
df_group_body
```

```python
df_pivot = df_group.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_pivot = df_pivot.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
```

```python
# Correlation
df.corr()
```

```python
pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
print(f"Pearson Correlation Coefficient: {pearson_coef}\nP-value: {p_value}")  
```
